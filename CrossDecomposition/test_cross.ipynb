```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.cross_decomposition import PLSRegression
import joblib

# Chargement des données et du modèle
X_test = np.load("X_test.npy")
y_test = np.load("y_test.npy")
pls = joblib.load("pls_model.joblib")

# Prédiction
Y_pred = pls.predict(X_test)
y_pred = np.argmax(Y_pred, axis=1)

# Test de précision
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy (PLSRegression): {acc:.4f}")

# Assertion simple (baseline de 0.6 par exemple)
assert acc > 0.6, "La précision du modèle est insuffisante"
print("Test réussi : le modèle atteint la précision attendue.")
```
